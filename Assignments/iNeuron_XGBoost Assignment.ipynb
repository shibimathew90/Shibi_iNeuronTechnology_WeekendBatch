```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
```

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)

test_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test', skiprows = 1, header = None)
```

```python
col_labels = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation','relationship', 'race', 'sex', 
              'capital_gain','capital_loss', 'hours_per_week', 'native_country', 'wage_class']
train_set.columns = col_labels
test_set.columns = col_labels
```

```python
train_set.head()
```

```python
train_set.shape
```

```python
# Combine both the set so that pre-processing can be done in same data frame.
train_set['Set']="train_set"
test_set['Set']="test_set"
data = train_set.append(test_set)
data.shape
```

```python
data.describe()
```

```python
data.describe(include='O')
# Use describe for object type columns
```

```python
data.info()
# Here we observe that there are no null values in any column
```

```python
obj_cols = []
num_cols = []
for col_type in data.columns:
    if data[col_type].dtype == 'object':
        obj_cols.append(col_type)
    else:
        num_cols.append(col_type)
```

```python
for col_item in obj_cols:
    print('_'*10,col_item, '_'*10)
    print(data[col_item].value_counts())
    print("Total Unique Count:",data[col_item].value_counts().count())
    print('_'*80)
```

```python
import re
```

```python
# Observe that in the wage class there is a '.' in column names. Need to clean Wage class column

data['wage_class'] = [re.sub(r"\.",'', x) for x in data['wage_class']]
data['wage_class'].unique()
```

```python
data['wage_class'].value_counts()
```

```python
le = LabelEncoder()
y = le.fit_transform(data['wage_class'])
# There are only two category(binary target) so we can apply LabelEncoder to convert it to numeric
```

```python
data2 = data.copy()
data3 = pd.get_dummies(data2[obj_cols[:-2]])
data3.head()
```

```python
# Observe that there are few columns which contain '?'. Lets remove those are we dont have any information related to these columns
for name in data3.columns:
    if(re.search(r'\?', name)):
        data3.drop(name, axis=1, inplace=True)
    
```

```python
data_fin = pd.concat([data3, data[num_cols], data['Set']], axis = 1)
```

```python
data_fin['Y'] = y
data_fin.head()
```

```python
data_train = data_fin[data_fin['Set'] == 'train_set']
data_test = data_fin[data_fin['Set'] == 'test_set']
```

```python
y_train=data_train['Y']
y_test=data_test['Y']

data_train.drop(['Set','Y'], axis=1, inplace=True)
data_test.drop(['Set','Y'], axis=1, inplace=True)


x_train= data_train.copy()
x_test = data_test.copy()
x_train.shape,x_test.shape, y_train.shape, y_test.shape
```

```python
from xgboost.sklearn import XGBClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score
```

```python
xgb = XGBClassifier()
xgb.fit(x_train, y_train)
```

```python
y_pred = xgb.predict(x_train)
accuracy = accuracy_score(y_train, y_pred)
accuracy
```

```python
y_pred = xgb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'learning_rate': [1, 0.5, 0.1, 0.01, 0.001],
    'max_depth': range(1, 20, 2),
    'n_estimators': [10,50,100,200],
    #'min_child_weight' : range(1, 8, 2),
    'gamma': [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
    
    }
```

```python
grid = GridSearchCV(XGBClassifier(objective='binary:logistic'), param_grid, verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred_new = new_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred_new)
accuracy
```
