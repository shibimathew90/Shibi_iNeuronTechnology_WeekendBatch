```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df['fg.'].fillna(df['fg.'].median(), inplace = True)
df['x3p.'].fillna(df['x3p.'].median(), inplace = True)
df['x2p.'].fillna(df['x2p.'].median(), inplace = True)
df['efg.'].fillna(df['efg.'].median(), inplace = True)
df['ft.'].fillna(df['ft.'].median(), inplace = True)
df.drop(['player', 'bref_team_id', 'season', 'season_end'], axis=1, inplace= True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
pos_new = pd.get_dummies(df['pos'], prefix= 'pos', drop_first= True)
df.drop('pos', axis = 1, inplace= True)
df = pd.concat([df, pos_new], axis = 1)
```

```python
df.head()
```

```python
X = df.drop(['pts'], axis = 1)
y = df['pts']
```

```python
y.head()
```

```python
X.head()
```

```python
normalized_nba_feature = preprocessing.normalize(X)
X_train, X_test, y_train, y_test = train_test_split(normalized_nba_feature, y, train_size=0.75, test_size=0.25, random_state = 100)
```

```python
error = []
for K in range(25):
    K_value = K+1
    neigh = KNeighborsClassifier(n_neighbors = K_value, weights='distance', algorithm='auto',metric='minkowski')
    neigh.fit(X_train, y_train)
    y_pred = neigh.predict(X_test)
    print("Accuracy is ", accuracy_score(y_test,y_pred)*1000,"% for K-Value:",K_value)
    error.append(np.mean(y_pred != y_test))
```

```python
# 66.11% accuracy on K = 3,11,12,13,14,15,16,17. A larger value of K will lead to greater amount of execution time & 
# hence underfitting. The small value of K will lead to overfitting. 
# So, K = 3.
```
